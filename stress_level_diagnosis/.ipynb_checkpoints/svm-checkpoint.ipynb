```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors

from sklearn.utils import resample
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import scale
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../split_train_test_data./stress_dataset.csv')
df.head()
```

```python
df.rename({
  'sr': 'SR',
  'rr': 'RR',
  't': 'T',
  'lm': 'LM',
  'bo': 'BO',
  'rem': 'REM',
  'sr.1': 'SR.1',
  'hr': 'HR',
  'sl': 'SL'
}, axis=1, inplace=True)
df.corr()
```

## Downsample the data

```python
df_sl_0 = df[df['SL'] == 0]
df_sl_1 = df[df['SL'] == 1]
df_sl_2 = df[df['SL'] == 2]
df_sl_3 = df[df['SL'] == 3]
df_sl_4 = df[df['SL'] == 4]
```

```python
arr = []
for col in df['SL'].unique():
  arr.append(resample(df[df['SL'] == col], replace=False, n_samples=100, random_state=42))
 
df_downsample = pd.concat(arr)
```

```python
X = df.drop('SL', axis=1).copy()
X.head()
```

```python
y = df['SL'].copy()
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=345)
X_train_scaled = scale(X_train)
X_test_scaled = scale(X_test)
```

## Membuat model Support Vector Machine

```python
clf_svm = SVC(random_state=42)
clf_svm.fit(X_train_scaled, y_train)
predictions = clf_svm.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, predictions, labels=clf_svm.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf_svm.classes_)

disp.plot()
plt.show()
```
